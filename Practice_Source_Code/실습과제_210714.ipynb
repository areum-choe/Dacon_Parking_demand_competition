```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
train=pd.read_csv("train_df_errno.csv")
test=pd.read_csv("test_df.csv")
```

```python
train.info()
```

```python
test.info()
```

```python
train.shape, test.shape
```

```python
sel = ['총세대수','전용면적별세대수','전용면적','공가수','단지내주차면수']

X = train[sel]
y = train['등록차량수']

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, y_train)   # 학습
pred = model.predict(X_test)
pred
```

```python
dict_dat = {"실제값 " : y_test, "예측값" : pred, "오차절대값" :abs( y_test - pred),
           "오차제곱" : (y_test - pred)**2}
dat = pd.DataFrame(dict_dat)
dat
```

```python
###학습 데이터 셋의 결정계수
print('학습 데이터 셋:',model.score(X_train,y_train))
print('테스트 데이터 셋:',model.score(X_test,y_test))
```

### MAE

```python
sum(dat['오차절대값'])/len(dat['오차절대값'])
```

### MSE

```python
mse_val = sum(dat['오차제곱']) / len(dat['오차제곱'])
mse_val
```

### RMSE 

```python
rmse = np.sqrt(mse_val)
rmse
```

### LASSO    

```python
from sklearn.linear_model import Lasso, Ridge # Lasso : L1, Ridge: L2
lasso = Lasso(alpha=0.01).fit(X_train, y_train)
print("학습용 데이터 세트 점수 : {:.2f}".format(lasso.score(X_train, y_train)))
print("테스트 데이터 세트 점수 : {:.2f}".format(lasso.score(X_test, y_test)))
```

### LIDGE

```python
from sklearn.linear_model import Lasso, Ridge # Lasso : L1, Ridge: L2
ridge = Ridge(alpha=0.1).fit(X_train, y_train)
print("학습용 데이터 세트 점수 : {:.2f}".format(ridge.score(X_train, y_train)))
print("테스트 데이터 세트 점수 : {:.2f}".format(ridge.score(X_test, y_test)))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

###  LASSO모델 적용해보기. MAE, MSE 구해보기

```python
sel = [0.0, 0.1, 0.2, 0.01, 0.3 ] #0.0001이 가장크다.
for var in sel:
    lasso = Lasso(alpha=var)
    lasso.fit(X_train, y_train)
    pred = lasso.predict(X_test)
    MAE = mean_absolute_error(y_test, pred)
    MSE = mean_squared_error(y_test, pred)
    print('alpha:{}'.format(var))
    print("학습용 데이터 세트 점수 : {:.3f}".format(lasso.score(X_train, y_train)))
    print("테스트 데이터 세트 점수 : {:.3f}".format(lasso.score(X_test, y_test)))
    print('MSE:{}'.format(MSE))
    print('MAE:{}'.format(MAE))
    print()
```

### LIDGE모델 적용해보기. MAE, MSE 구해보기

```python
sel = [0.0, 0.1, 0.2, 0.01, 0.3 ] #0.0001이 가장크다.
for var in sel:
    ridge = Ridge(alpha=var)
    ridge.fit(X_train, y_train)
    pred = ridge.predict(X_test)
    MAE = mean_absolute_error(y_test, pred)
    MSE = mean_squared_error(y_test, pred)
    print('alpha:{}'.format(var))
    print("학습용 데이터 세트 점수 : {:.3f}".format(ridge.score(X_train, y_train)))
    print("테스트 데이터 세트 점수 : {:.3f}".format(ridge.score(X_test, y_test)))
    print('MSE:{}'.format(MSE))
    print('MAE:{}'.format(MAE))
    print()
```
